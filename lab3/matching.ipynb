```python
import cv2
import matplotlib.pyplot as plt
```

```python
ORimage = cv2.imread('landscape_2.jpg')
ORimage = cv2.cvtColor(ORimage, cv2.COLOR_BGR2RGB)

height, width, _ = ORimage.shape

   
third_width = (width // 3)


file_names = [ORimage[:, :third_width+100], ORimage[:, third_width-300:2*third_width+300], ORimage[:, 100-2*third_width:]]




#file_names = ['1.jpg','2.jpg','3.jpg']

images = []
for file_name in file_names:
    #image = cv2.imread(file_name)
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image=file_name
    plt.figure(figsize=(5,5))
    plt.imshow(image)

    images.append(image)
```

```python
# Remove the last element
print(len(images))
images.pop(-1)
print(len(images))

stitchy=cv2.Stitcher.create()
(status,output)=stitchy.stitch(images)

# Check if the sticking procedure was successful
if status == cv2.STITCHER_OK:
  stitching_completed = True
  plt.imshow(output)
else:
  print("There was an error stitching the images")
```
