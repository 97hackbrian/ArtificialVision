Una vez terminado el entrenamiento se realizaran algunas pruebas de deteccion con la camara

```python
# Importamos las librerias
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import time

# Leer nuestro modelo
model = YOLO("../models/modelV1.pt")

# Realizar VideoCaptura
cap = cv2.VideoCapture(0)

# Configurar la frecuencia de captura (cada 3 segundos)
intervalo_captura = 5  # en segundos
ultimo_tiempo_captura = time.time()

# Bucle
while True:
    # Leer nuestros fotogramas
    ret, frame = cap.read()

    # Leemos resultados
    resultados = model.predict(frame, imgsz=640, conf=0.6)

    # Mostramos resultados
    anotaciones = resultados[0].plot()

    # Mostramos nuestros fotogramas
    cv2.imshow("DETECCION", anotaciones)

    # Capturar imagen cada intervalo_captura segundos
    tiempo_actual = time.time()
    if tiempo_actual - ultimo_tiempo_captura >= intervalo_captura:
        # Guardar la imagen capturada
        nombre_archivo = f"imagen_{int(tiempo_actual)}.jpg"
        cv2.imwrite(nombre_archivo, frame)
        print(f"Imagen guardada como {nombre_archivo}")

        # Actualizar el tiempo de la √∫ltima captura
        ultimo_tiempo_captura = tiempo_actual

        # Procesar la imagen guardada por el modelo
        imagen_guardada = cv2.imread(nombre_archivo)
        resultados_guardados = model.predict(imagen_guardada, imgsz=640, conf=0.7)

        # Mostrar resultados de la imagen guardada
        anotaciones_guardadas = resultados_guardados[0].plot()
        plt.imshow(cv2.cvtColor(anotaciones_guardadas, cv2.COLOR_BGR2RGB))
        plt.title("Resultados de la imagen guardada")
        plt.show()

    # Cerrar nuestro programa
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
