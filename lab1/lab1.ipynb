```python
imgCv2=cv2.imread('powerrangers.jpeg')
plt.imshow(imgCv2)
```

```python
image2CV2convert = cv2.cvtColor(imgCv2,cv2.COLOR_BGR2RGB)

print(type(image2CV2convert))
print(image2CV2convert.shape)
print(image2CV2convert.size)
print(image2CV2convert.dtype)
plt.imshow(image2CV2convert)
```

```python
image2CV2convert = cv2.cvtColor(imgCv2,cv2.COLOR_BGR2RGB)
image2CV2convert2 = cv2.cvtColor(imgCv2,cv2.COLOR_BGR2GRAY)
print(image2CV2convert2)
print(image2CV2convert)
plt.imshow(image2CV2convert,cmap="winter")
plt.imshow(image2CV2convert2,cmap="winter",alpha=0.5)
```

```python
image2CV2convert2[15:60,150:187]=0
print(image2CV2convert2.shape)
plt.imshow(image2CV2convert2,cmap="gray")
```

```python
image2CV2convertx=image2CV2convert
image2CV2convertx[15:60,150:187,:]=[0,0,0]
print(image2CV2convertx.shape)
plt.imshow(image2CV2convertx)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('normal')
ax1.imshow(image2CV2convert)
ax2.set_title('gray')
ax2.imshow(image2CV2convert2, cmap='gray')
ax3.set_title('slicing')
ax3.imshow(image2CV2convertx)
```
